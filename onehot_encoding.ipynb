```python
import pandas as pd

df = pd.DataFrame({'employee':["abc","def","ghi","jkl","mno","pqr"],'age':[34,27,27,26,21,32],'city':["ahemadabad","banglore","cochi","ahemadabad","ahemadabad","cochi"]})
df
```

```python
from sklearn.preprocessing import LabelEncoder,LabelBinarizer
le=LabelEncoder()
lb=LabelBinarizer()
df["city_label"]=le.fit_transform(df["city"])
df
```

```python
city_one_hot=lb.fit_transform(df["city_label"])
```

```python
pd.DataFrame(city_one_hot)
```

```python
pd.concat([df,pd.DataFrame(city_one_hot)],axis=1)
```
